```python
#working with dataset
import pandas as pd
```

```python
pubg=pd.read_csv('PlayerUnknowns Battlegrounds_GunInfo.csv')
df_pubg=pubg.apply(pd.to_numeric,errors='ignore')
df_pubg.head(10)
```

```python
trace=go.Scatter(x=df_pubg['ammunition'],y=df_pubg['1hurthead'],
       name='category')
layout=go.Layout(title='PUBG ammo vs death',
       plot_bgcolor='rgb(230,230,230)',
       showlegend=True)
fig=go.Figure(data=[trace],layout=layout)
iplot(fig, filename='tset')
```

```python
#bar plot
trace1=go.Bar(
    x=df_pubg['category'],
    y=df_pubg['1hurt'],
    name='1 damage'
    )
trace2=go.Bar(
    x=df_pubg['category'],
    y=df_pubg['3hurt'],
    name='2 damage')
```

```python
data=[trace1,trace2]
layout=go.Layout(barmode='group')
fig=go.Figure(data=data, layout=layout)
iplot(fig, filename='grouped-bar')
```

```python
#3D plotly visualization
```

```python
x=df_pubg['ammunition']
y=df_pubg['volume']
z=df_pubg['1hurt']

trace1=go.Scatter3d(
    x=x,y=y,z=z,mode='markers',marker=dict(size=12,color=z,colorscale='Viridis',opacity=0.8))

data=[trace1]
layout=go.Layout(margin=dict(
    l=0,r=0,b=0,t=0   #margin in px
))

fig=go.Figure(data=data, layout=layout)
iplot(fig, filename='PlayerUnknowns Battlegrounds_GunInfo')

```

```python
#plotting on apple stock data
apple=pd.read_csv('finance-charts-apple.csv')

trace_one=go.Scatter(
    x=apple['Date'],
    y=apple['AAPL.High'],
    name="Apple High",
    line=dict(color='#17BECF'),
    opacity=0.8
)

trace_two=go.Scatter(
    x=apple['Date'],
    y=apple['AAPL.Low'],
    name="apple LOW",
    line=dict(color='#7F7F7F'),
    opacity=0.8
)
data=[trace_one, trace_two]

layout=dict(
    title='Apple Stock price- high vs Low')
fig=dict(data=data, layout=layout)
iplot(fig, filename='Apple Stock Comparision')
```

```python
import plotly.figure_factory as ff

scope=['Washington']
df=pd.read_csv('minoritymajority.csv')
```

```python
df_new=df[df['STNAME'].isin(scope)]

values=df_new['TOT_POP'].tolist()
fips=df_new['FIPS'].tolist()
```

```python
colorscale=["#36b431", "#c26f04", "#a434e3", "#1d3059", "#9cb850", "#5eaec9", "#be6bf4", "#7f17bc",
         "#aaed24", "#8dcfe6", "#7b7e10", "#91ca4e", "#ba4c93", "#30cd5a", "#efd557", "#7e8fea",
           "#5083c9", "#6c6678", "#eac0fe", "#926eb5", "#670ac0", "#432912", "#fccc32", "#8233ee",
             "#259a27", "#fdd623", "#a1d071", "#403e3c", "#e15eed", "#4ba1a0", "#fb4b57", "#efd7d0",
               "#c76ba1", "#9c5b91", "#d5bd9b", "#820582", "#448188", "#04864f", "#fce0ac", "#7bfc69",
                 "#a83dab", "#1e556d", "#cbf84d", "#0ef476", "#6684f6", "#fc045b", "#9b8bdd", "#2abb14", 
                 "#6cb8ba", "#014e50", "#dda473", "#b727c7", "#f3713e", "#9ca83a", "#869506", "#ad7af5",
                   "#305ecd", "#e04a41", "#0a6f92", "#ee21a9", "#0386f3", "#e98d2a", "#9d1be9","#7f7f7f"]
```

```python

fig=ff.create_choropleth(
    fips=fips,values=values, scope=scope,
    colorscale=colorscale,round_legend_values=True,
    simplify_county=0,
    simplify_state=0,
    county_outline={'color':'rgb(15,15,55)','width':0.5},
    state_outline={'width':0.5},
    legend_title='Population Per Country',
    title='Washington'
)

iplot(fig, filename='Cholorpleth Map Creation')
```
